```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from __future__ import division 
from scipy import *                  
from pylab import * 
import os

UA=149597870700
jour=3600*24
annee=365*3600*24
tau=2*np.pi

#Rayons des orbites des planÃ¨tes
Rmercure=0.38709893*UA
Rvenus=0.723332*UA
Rterre=1*UA
Rmars=1.52366231*UA
Rjupiter=5.20336301*UA
Rsoleil=696342000

#PÃ©riodes des planÃ¨tes
Tmercure=87.96934*jour
Tvenus=224.701*jour
Tterre=1*annee
Tmars=1.8808*annee
Tjupiter=11.862*annee
Tsoleil=27.28*jour


Lplanetes=[(Rmercure,Tmercure),(Rvenus,Tvenus),(Rterre,Tterre),(Rmars,Tmars),(Rjupiter,Tjupiter)]

#DurÃ©e de la simulation
TT=Tjupiter
#Pas de temps de la simulation
pas=1*jour

#angle continu
def theta(T):
    return (tau*np.linspace(0, TT, TT/pas)/T)%tau

#Listes des coordonnÃ©es des planÃ¨tes du systÃ¨me solaire interne
#(Mercure, VÃ©nus, Terre, Mars, Jupiter)
Lint=[(Rmercure,Tmercure),(Rvenus,Tvenus),(Rterre,Tterre),(Rmars,Tmars),(Rjupiter,Tjupiter)]
Xint=[R*np.cos(theta(T)) for (R,T) in Lint]
Yint=[R*np.sin(theta(T)) for (R,T) in Lint]
couleursint=['grey','green','blue','red','orange']
nomsint=['Mercure','VÃ©nus','Terre','Mars','Jupiter']

#TracÃ© des orbites internes
for (x,y,c,n) in zip(Xint,Yint,couleursint,nomsint):
    plt.plot(x,y,color=c,linewidth=1,linestyle='-',label=n)
plt.plot(0,0,'o',color='orange')
plt.axis("equal")
xticks([2*n*UA for n in range(-3,4)],[str(2*n)+' UA' for n in range(-3,4)])
yticks([n*UA for n in range(-6,7)],[str(n)+' UA' for n in range(-6,7)])
plt.title("Orbites des planÃ¨tes du systÃ¨me solaire interne")
legend(loc='upper left')
plt.show
```

```python
#mvt des planÃ¨tes dans le rÃ©fÃ©rentiel gÃ©ocentrique dans le systeme solaire interne
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt


#Listes des coordonnÃ©es des planÃ¨tes  du systÃ¨me solaire interne dans le rÃ©fÃ©rentiel gÃ©ocentrique
Xgeoint=[R*np.cos(theta(T)) for (R,T) in Lplanetes[:5]]
Ygeoint=[R*np.sin(theta(T)) for (R,T) in Lplanetes[:5]]
xtgeoint=Xgeoint[2]
ytgeoint=Ygeoint[2]

#TracÃ© des trajectoires de Mercure, VÃ©nus, Mars, Jupiter dans le rÃ©fÃ©rentiel gÃ©ocentrique
for (x,y,c,n) in zip(Xgeoint,Ygeoint,couleursint,nomsint):
    plt.plot(x-xtgeoint,y-ytgeoint,color=c,linewidth=1,linestyle='-',label=n)
plt.axis("equal")
xticks([2*n*UA for n in range(-3,4)],[str(2*n)+' UA' for n in range(-3,4)])
yticks([n*UA for n in range(-6,7)],[str(n)+' UA' for n in range(-6,7)])
plt.title("Trajectoires du systÃ¨me solaire interne dans le rÃ©fÃ©rentiel gÃ©ocentrique")
legend(loc='upper left')
plt.show
```
