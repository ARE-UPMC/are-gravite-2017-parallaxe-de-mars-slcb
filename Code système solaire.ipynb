```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from __future__ import division 
from scipy import *                  
from pylab import * 
import os

UA=149597870700
jour=3600*24
annee=365*3600*24
k=2*np.pi

#Rayons des orbites des planÃ¨tes
Rmercure=0.38709893*UA
Rvenus=0.723332*UA
Rterre=1*UA
Rmars=1.52366231*UA
Rjupiter=5.20336301*UA
Rsaturne=9.554909*UA
Ruranus=19.189165*UA
Rneptune=30.10366151*UA
Rsoleil=696342000

#PÃ©riodes des planÃ¨tes
Tmercure=87.96934*jour
Tvenus=224.701*jour
Tterre=1*annee
Tmars=1.8808*annee
Tjupiter=11.862*annee
Tsaturne=10757.7365*jour
Turanus=84.016846*annee
Tneptune=60224.9036*jour
Tsoleil=27.28*jour

Lplanetes=[(Rmercure,Tmercure),(Rvenus,Tvenus),(Rterre,Tterre),(Rmars,Tmars),(Rjupiter,Tjupiter),
   (Rsaturne,Tsaturne),(Ruranus,Turanus),(Rneptune,Tneptune)]

#angle continu
def theta(T):
    return k*np.linspace(0, T, 1000)/T

#Listes des coordonnÃ©es des planÃ¨tes
X=[R*np.cos(theta(T)) for (R,T) in Lplanetes]
Y=[R*np.sin(theta(T)) for (R,T) in Lplanetes]
#Couleurs des orbites
couleurs=['grey','green','blue','red','orange','black','green','blue']
noms=['Mercure','VÃ©nus','Terre','Mars','Jupiter','Saturne','Uranus','Neptune']

#CoordonnÃ©es des planÃ¨tes
#Soleil
xsol=Rsoleil*np.cos(theta(Tsoleil))
ysol=Rsoleil*np.sin(theta(Tsoleil))
#Mercure
xme=X[0]
yme=Y[0]
#VÃ©nus
xv=X[1]
yv=Y[1]
#Terre
xt=X[2]
yt=Y[2]
#Mars
xma=X[3]
yma=Y[3]
#Jupiter
xj=X[4]
yj=Y[4]
#Saturne
xsa=X[5]
ysa=Y[5]
#Uranus
xu=X[6]
yu=Y[6]
#Neptune
xn=X[7]
yn=Y[7]

#TracÃ© des orbites
for (x,y,c,n) in zip(X,Y,couleurs,noms):
    plt.plot(x,y,color=c,linewidth=1,linestyle='-',label=n)
plt.plot(xsol,ysol,color='orange',linewidth=5,linestyle='-')
plt.axis("equal")
plt.title("Orbites des planÃ¨tes autour du Soleil")
legend(loc='upper left')
plt.show
```

```python
#mvt des planÃ¨tes dans le rÃ©fÃ©rentiel gÃ©ocentrique
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt

#angle continu
def thetageo(T):
    return k*np.linspace(0, Tneptune, 1000)/T

#Listes des coordonnÃ©es des planÃ¨tes dans le rÃ©fÃ©rentiel gÃ©ocentrique
Xgeo=[R*np.cos(thetageo(T)) for (R,T) in Lplanetes]
Ygeo=[R*np.sin(thetageo(T)) for (R,T) in Lplanetes]
xtgeo=Xgeo[2]
ytgeo=Ygeo[2]

#TracÃ© des trajectoires dans le rÃ©fÃ©rentiel gÃ©ocentrique
for (x,y,c,n) in zip(Xgeo,Ygeo,couleurs,noms):
    plt.plot(x-xtgeo,y-ytgeo,color=c,linewidth=1,linestyle='-',label=n)
plt.axis("equal")
plt.title("Trajectoires des planÃ¨tes dans le rÃ©fÃ©rentiel gÃ©ocentrique")
legend(loc='upper left')
plt.show
```
